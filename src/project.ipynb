Data set was documented in previous assignment

1 type conversion, 1 calculate summary statistics, 1 calculate value counts, 1 visualization were performed

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/raw/NYPD_Hate_Crimes.csv', header=0)
df.info()
```

convert arrest ID to a string (type covnersion)

get value counts by type of offense in the race/color category


```python
df['Arrest Id'] = df['Arrest Id'].astype(str)
value_counts = df[df['Offense Category'] == 'Race/Color']['Offense Description'].value_counts()
print("Offense Description of Race/Color hate crimes:\n", value_counts)
```

summary stats

```python
summary_stats = df['Complaint Precinct Code'].describe()
print(summary_stats)
```

visualization of all hate crimes by bias type in NYC


```python
bias_counts = df['Bias Motive Description'].value_counts()
bias_counts.plot.bar()
plt.title('NYC Hate Crimes by Bias Type')
plt.xlabel('Bias Type')
plt.ylabel('Number of Crimes')
plt.show()
```
